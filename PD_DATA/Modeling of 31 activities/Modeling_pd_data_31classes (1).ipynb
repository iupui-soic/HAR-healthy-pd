```python
# convert to numerical
df = df.astype('category')
column = df.select_dtypes(['category']).columns
df[column] = df[column].apply(lambda x: x.cat.codes)
display(df)
```

```python
#scaling the data
from sklearn.preprocessing import MinMaxScaler
Scaler= MinMaxScaler()
df['Accel X (g)_hand']=Scaler.fit_transform(df['Accel X (g)_hand'].values.reshape(-1, 1))
df['Accel Y (g)_hand']=Scaler.fit_transform(df['Accel Y (g)_hand'].values.reshape(-1, 1))
df['Accel Z (g)_hand']=Scaler.fit_transform(df['Accel Z (g)_hand'].values.reshape(-1, 1))
df['Gyro X (°/s)_hand']=Scaler.fit_transform(df['Gyro X (°/s)_hand'].values.reshape(-1, 1))
df['Gyro Y (°/s)_hand']=Scaler.fit_transform(df['Gyro Y (°/s)_hand'].values.reshape(-1, 1))
df['Gyro Z (°/s)_hand']=Scaler.fit_transform(df['Gyro Z (°/s)_hand'].values.reshape(-1, 1))

df['Accel X (g)_leg']=Scaler.fit_transform(df['Accel X (g)_leg'].values.reshape(-1, 1))
df['Accel Y (g)_leg']=Scaler.fit_transform(df['Accel Y (g)_leg'].values.reshape(-1, 1))
df['Accel Z (g)_leg']=Scaler.fit_transform(df['Accel Z (g)_leg'].values.reshape(-1, 1))
df['Gyro X (°/s)_leg']=Scaler.fit_transform(df['Gyro X (°/s)_leg'].values.reshape(-1, 1))
df['Gyro Y (°/s)_leg']=Scaler.fit_transform(df['Gyro Y (°/s)_leg'].values.reshape(-1, 1))
df['Gyro Z (°/s)_leg']=Scaler.fit_transform(df['Gyro Z (°/s)_leg'].values.reshape(-1, 1))

df['Accel X (g)_chest']=Scaler.fit_transform(df['Accel X (g)_chest'].values.reshape(-1, 1))
df['Accel Y (g)_chest']=Scaler.fit_transform(df['Accel Y (g)_chest'].values.reshape(-1, 1))
df['Accel Z (g)_chest']=Scaler.fit_transform(df['Accel Z (g)_chest'].values.reshape(-1, 1))
df['Sample (V)']=Scaler.fit_transform(df['Sample (V)'].values.reshape(-1, 1))
```

```python
from sklearn.model_selection import train_test_split
y = df['Activity_Label']
X = df.drop(['Activity_Label'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)
```

```python
print('X_train.shape',X_train.shape)
print('y_train.shape',y_train.shape)
print('X_test.shape',X_test.shape)
```

```python
#rfc
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)

rfc_predicted = rfc.predict(X_test)
rfc_expected = y_test
rfc_y_proba = rfc.predict_proba(X_test)[:,1]
score = rfc.score(X_test, y_test)
print('RFC =',score)
train_acc = rfc.score(X_train, y_train)
print("The Training Accuracy is {}".format(train_acc*100))
from sklearn.metrics import classification_report
print(classification_report(rfc_expected, rfc_predicted))
```

```python
#XGB
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)

xgb_predicted = xgb.predict(X_test)
xgb_expected = y_test
xgb_y_proba = xgb.predict_proba(X_test)[:,1]
score = xgb.score(X_test, y_test)
print('XGB =',score)
train_acc = xgb.score(X_train, y_train)
print("The Training Accuracy is {}".format(train_acc*100))
from sklearn.metrics import classification_report
print(classification_report(xgb_expected, xgb_predicted))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler


from scipy import stats
%matplotlib inline
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential, layers, callbacks
from tensorflow.keras.layers import Dense, LSTM, Dropout, GRU, Bidirectional
```

```python
#BiLSTM_PD
from sklearn.preprocessing import RobustScaler
import datetime
```

```python
df1 = pd.read_csv('PDdata_31activities.csv')
df1_train = df1.sample(frac=0.7)
df1_test = df1.loc[~df1.index.isin(df1_train.index)]
```

```python
#scaling with robust scaler
scale_columns = ['Accel X (g)_hand', 'Accel Y (g)_hand', 'Accel Z (g)_hand','Gyro X (°/s)_hand','Gyro Y (°/s)_hand','Gyro Z (°/s)_hand', 'Accel X (g)_leg', 'Accel Y (g)_leg', 'Accel Z (g)_leg','Gyro X (°/s)_leg','Gyro Y (°/s)_leg','Gyro Z (°/s)_leg','Accel X (g)_chest', 'Accel Y (g)_chest', 'Accel Z (g)_chest','Sample (V)']

scaler = RobustScaler()

scaler = scaler.fit(df1_train[scale_columns])

df1_train.loc[:, scale_columns] = scaler.transform(
  df1_train[scale_columns].to_numpy()
)

df1_test.loc[:, scale_columns] = scaler.transform(
  df1_test[scale_columns].to_numpy()
)
```

```python
def create_dataset(X, y, time_steps=1, step=1):
    Xs, ys = [], []
    for i in range(0, len(X) - time_steps, step):
        v = X.iloc[i:(i + time_steps)].values
        labels = y.iloc[i: i + time_steps]
        Xs.append(v)
        ys.append(stats.mode(labels)[0][0])
    return np.array(Xs), np.array(ys).reshape(-1, 1)
```

```python
TIME_STEPS = 200
STEP = 20

X_train, y_train = create_dataset(
    df1_train[['Accel X (g)_hand', 'Accel Y (g)_hand', 'Accel Z (g)_hand','Gyro X (°/s)_hand','Gyro Y (°/s)_hand','Gyro Z (°/s)_hand', 'Accel X (g)_leg', 'Accel Y (g)_leg', 'Accel Z (g)_leg','Gyro X (°/s)_leg','Gyro Y (°/s)_leg','Gyro Z (°/s)_leg','Accel X (g)_chest', 'Accel Y (g)_chest', 'Accel Z (g)_chest','Sample (V)']],
    df1_train.Activity_Label,
    TIME_STEPS,
    STEP
)

X_test, y_test = create_dataset(
    df1_test[['Accel X (g)_hand', 'Accel Y (g)_hand', 'Accel Z (g)_hand','Gyro X (°/s)_hand','Gyro Y (°/s)_hand','Gyro Z (°/s)_hand', 'Accel X (g)_leg', 'Accel Y (g)_leg', 'Accel Z (g)_leg','Gyro X (°/s)_leg','Gyro Y (°/s)_leg','Gyro Z (°/s)_leg','Accel X (g)_chest', 'Accel Y (g)_chest', 'Accel Z (g)_chest','Sample (V)']],
    df1_test.Activity_Label,
    TIME_STEPS,
    STEP
)
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore', sparse=False)

enc = enc.fit(y_train)

y_train = enc.transform(y_train)
y_test = enc.transform(y_test)
```

```python
model = keras.Sequential()
model.add(
    keras.layers.Bidirectional(
      keras.layers.LSTM(
          units=128,
          input_shape=[X_train.shape[1], X_train.shape[2]]
      )
    )
)
model.add(keras.layers.Dropout(rate=0.3))
model.add(keras.layers.Dense(units=64, activation='relu'))
model.add(keras.layers.Dropout(rate=0.2))
model.add(keras.layers.Dense(units=32, activation='relu'))
model.add(keras.layers.Dense(units=6, activation='softmax'))
model.add(keras.layers.Dense(y_train.shape[1], activation='softmax'))

model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['acc']
)
```

```python

history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.1,
    shuffle=False
)
```

```python
# evaluate the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=2)
print('BiLSTM Accuracy: %f' % (accuracy*100))
```
