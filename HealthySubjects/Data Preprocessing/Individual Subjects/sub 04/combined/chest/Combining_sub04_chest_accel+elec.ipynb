```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
accel = pd.read_csv('final_cleaned_chest_accel-sub04.csv-sub04')
elec = pd.read_csv('final_cleaned_chest_elec-sub04.csv-sub04')
```

```python
elec_ts = elec['Timestamp (microseconds)'].to_numpy()
```

```python
elec_ts
```

```python
accel_ts = accel['Timestamp (microseconds)'].to_numpy()
```

```python
accel_ts
```

```python
for i in range(len(accel_ts)):
    accel.at[i, 'Sample (V)'] = elec['Sample (V)'].iloc[np.argmin(abs(accel_ts[i] - elec_ts))]
```

```python
accel
```

```python
ts = accel['Timestamp (microseconds)'].to_numpy()
ts1 = ((ts - ts[0])*1e-6)
```

```python
accel['Time_elapse'] = ts1.tolist()
```
